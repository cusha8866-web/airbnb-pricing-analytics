```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Airbnb_Data.csv", low_memory = False)
```

```python
df
```

```python
df.columns
```

# Checking Missing Values


```python
print(df.isnull().sum())
```

```python
df.info()
```

# Handling Missing Values

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month':0, 'last review' : df['last review'].min()}, inplace = True)
```

```python
df.dropna(subset = ['NAME', 'host name'], inplace =True)
```

```python
print(df.isnull().sum())
```

# Removing Unwanted Columns

```python
df = df.drop(columns=['license', 'house_rules'], errors= 'ignore')
```

```python
df.head()
```

```python
# Remove Dollar Signs And Convert to Float
df['price'] = df['price'].replace(r'[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

# Descriptive Statistics

```python
df.describe()
```

# Visualization

# What is the distrubution of listing prices?

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=50, kde = True, color = 'Green')
plt.title("Distribution of Listing Price")
plt.xlabel("Price($)")
plt.ylabel("Frequency")
plt.show()
```

# How are different room types distributed?

```python
df['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot( x = "room type", data = df, color ="blue")
plt.xlabel("Room Type")
plt.ylabel("No. of Rooms")
plt.title("Distrubution of Room Type")
plt.show()
```

# How are listings distirbuted acorss different neighborhoods?

```python
plt.figure(figsize=(9,6))
sns.countplot(y='neighbourhood group', data = df, color = 'yellow', order= df['neighbourhood group'].value_counts().index)
plt.title("No. of Listings by Neighbourhood Groups")
plt.xlabel(" No. of Listings")
plt.ylabel("Neighbourhood Groups")
plt.show()
```

# What is the relationship between Price and Room Type?

```python
plt.figure(figsize=(8,6))
sns.boxplot(x = 'room type' , y = 'price' ,hue ='room type', data = df, palette = 'Set1' )
plt.title("Price vs Room Type")
plt.xlabel("Room Type")
plt.ylabel("Price ($)")
plt.legend(title = 'Room Type')
plt.show()
```

# How has the number of reviews change over with time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12,6))
reviews_over_time.plot(kind = 'line' , color = 'red')
plt.title("Number of Reviews Over Time")
plt.xlabel(" Date")
plt.ylabel("Number of Reviews")
plt.show()
```
